```python
import os
import pandas as pd
import numpy as np
import anndata as ann

import spatialdm as sdm
from spatialdm.datasets import dataset
import spatialdm.plottings as pl

import matplotlib.pyplot as plt
print("SpatailDM version: %s" %sdm.__version__)

from scipy.stats import spearmanr, pearsonr
import scanpy as sc
import liana as li
```

```python
adata = dataset.melanoma()
```

```python
# NOTE: SpatialDM uses adata for global, and adata.raw for local -> make them identical
# I strongly disagree with permuting raw counts,
adata = adata.raw.to_adata()
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
adata.raw = adata.copy()
```

```python
sdm.weight_matrix(adata=adata, l=1.2, cutoff=0.1, single_cell=False) # weight_matrix by rbf kernel
```

```python
# visualize range of interaction
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3))

# plt.scatter(adata.obsm['spatial'][:,0], adata.obsm['spatial'][:,1],
#             c=adata.obsp['weight'][50])

plt.scatter(adata.obsm['spatial'][:,0], 
            adata.obsm['spatial'][:,1],
            c=adata.obsp['weight'].A[50])

```

```python
import time
start = time.time()
sdm.weight_matrix(adata, l=1.2, cutoff=0.2, single_cell=False) # weight_matrix by rbf kernel
sdm.extract_lr(adata, 'human', min_cell=3)
```

```python
adata.obsp['nearest_neighbors'] = adata.obsp['weight']
```

```python
sdm.spatialdm_global(adata, 100, specified_ind=None, method='both', nproc=1)     # global Moran selection
sdm.sig_pairs(adata, method='permutation', fdr=False, threshold=2)     # select significant pairs
sdm.spatialdm_local(adata, n_perm=100, method='both', specified_ind=None, nproc=1, scale_X=False)     # local spot selection
```

```python
adata.uns.keys()
```

```python
def _uns(adata, key):
    return adata.uns[key]
```

```python
adata
```

```python
global_res = adata.uns['global_res']
global_res['morans_r'] = adata.uns['global_I'].flatten()
```

```python
local_i = _uns(adata, 'local_stat')['local_I']
local_i_r = _uns(adata, 'local_stat')['local_I_R']
local_r = local_i + local_i_r
local_msk = ((local_i > 0) * (local_i_r > 0)).astype(int)
local_p = adata.uns['local_perm_p']
```

```python
spdata = ann.AnnData(X=local_r,
                     obs=adata.obs,
                     var=global_res,
                     obsm=adata.obsm,
                     layers={'pvals':local_p.T, 'cats':local_msk})
```

```python
li.ut.spatial_neighbors(adata, bandwidth=1.2, cutoff=0.2, set_diag=True)
```

```python
li.mt.lr_bivar(adata, 
               function_name="morans",
               inplace=True,
               use_raw=True, # NOTE 
               n_perms=100,
               expr_prop=0.1,
               verbose=True,
               mask_negatives=True,
               resource_name='cellchatdb',
               connectivity_key='weight',
               add_categories=True
               )
```
